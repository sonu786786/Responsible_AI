```python
import zipfile

with zipfile.ZipFile("salary.zip", "r") as zip_ref:
    zip_ref.extractall()
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
df = pd.read_csv("Salary_dataset.csv")
df.head()
```

```python
df.info()
```

```python
X = df.iloc[:, :-1]   # all columns except last
y = df.iloc[:, -1]    # last column
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.30, shuffle=False
)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)

mse = mean_squared_error(y_test, y_pred)
mse
```

```python
mse_values = []

for i in range(5):
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=0.30, shuffle=True, random_state=i
    )

    model = LinearRegression()
    model.fit(X_train, y_train)

    y_pred = model.predict(X_test)
    mse = mean_squared_error(y_test, y_pred)

    mse_values.append(mse)

    print(f"Shuffle {i+1} - MSE: {mse}")
```

```python
mse_values
```

Conculusion:                     When the dataset is split without shuffling, the model performance depends on the original order of the data.
After shuffling the data multiple times, the Mean Squared Error changes slightly due to different distributions of training and testing samples.
This shows that shuffling has an impact on model performance and helps achieve a more generalized model.
